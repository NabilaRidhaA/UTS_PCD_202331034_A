## 202331034_Nabila Ridha Adman

```python
image_path = 'gambar asli.png'
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

## 202331034_Nabila Ridha Adman

```python
height, width = img.shape[:2]
```

## 202331034_Nabila Ridha Adman

```python
grayscale = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

## 202331034_Nabila Ridha Adman

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(grayscale, cmap='gray')
plt.title("Gambar Grayscale Awal")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(grayscale.ravel(), bins=256, range=(0, 256), color='blue', alpha=0.7)
plt.title("Histogram Gambar Grayscale Awal")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi")

plt.tight_layout()
plt.show()
```

## 202331034_Nabila Ridha Adman

```python
threshold = 127
binary_image = np.where(grayscale > threshold, 255, 0).astype(np.uint8)

# 3. Tampilkan hasil
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(grayscale, cmap='gray')
axes[0].set_title("Grayscale")
axes[0].axis('off')

axes[1].imshow(binary_image, cmap='gray')
axes[1].set_title("Binary")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

## 202331034_Nabila Ridha Adman

```python
beta = 40 
citra_cerah = np.zeros((height, width), dtype=np.uint8)

for x in range(height):
    for y in range(width):
        pixel_value = grayscale[x, y] + beta
        if pixel_value > 255:
            pixel_value = 255
        citra_cerah[x, y] = pixel_value
        
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(citra_cerah, cmap='gray')
plt.title("Gambar Dipercerah")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(citra_cerah.ravel(), bins=256, range=(0, 256), color='red', alpha=0.7)
plt.title("Histogram Gambar Dipercerah")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi")

plt.tight_layout()
plt.show()
```

## 202331034_Nabila Ridha Adman

```python
alpha = 1.2 
citra_kontras = np.zeros((height, width), dtype=np.uint8)

for x in range(height):
    for y in range(width):
        pixel_value = grayscale[x, y] * alpha
        if pixel_value > 255:
            pixel_value = 255
        citra_kontras[x, y] = pixel_value

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(citra_kontras, cmap='gray')
plt.title("Gambar Diperkontras")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(citra_kontras.ravel(), bins=256, range=(0, 256), color='green', alpha=0.7)
plt.title("Histogram Gambar Diperkontras")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi")

plt.tight_layout()
plt.show()
```

## 202331034_Nabila Ridha Adman

```python
beta = 10  
alpha = 1.2  
citra_hasil = np.zeros((height, width), dtype=np.uint8)

for x in range(height):
    for y in range(width):
        pixel_value = alpha * grayscale[x, y] + beta
        if pixel_value > 255:
            pixel_value = 255
        citra_hasil[x, y] = pixel_value

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(citra_hasil, cmap='gray')
plt.title("Gambar Dipercerah dan Diperkontras")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(citra_hasil.ravel(), bins=256, range=(0, 256), color='orange', alpha=0.7)
plt.title("Histogram Gambar Dipercerah dan Diperkontras")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi")

plt.tight_layout()
plt.show()
```

## 202331034_Nabila Ridha Adman

```python
beta = 10
alpha = 1.2
citra_negatif = np.zeros((height, width), dtype=np.uint8)

for x in range(height):
    for y in range(width):
        pixel_value = 255 - grayscale[x, y]
        citra_negatif[x, y] = pixel_value

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(citra_negatif, cmap='gray')
plt.title("Gambar Negatif")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.hist(citra_negatif.ravel(), bins=256, range=(0, 256), color='purple', alpha=0.7)
plt.title("Histogram Gambar Negatif")
plt.xlabel("Intensitas Pixel")
plt.ylabel("Frekuensi")

plt.tight_layout()
plt.show()
```
