```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 202331034_Nabila Ridha Adman

```python
img = cv2.imread('nama.png')
```

## 202331034_Nabila Ridha Adman

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

## 202331034_Nabila Ridha Adman

```python
mask_blue = cv2.inRange(hsv, (100, 50, 50), (140, 255, 255))
mask_green = cv2.inRange(hsv, (30, 40, 40), (100, 255, 255))
mask_red = cv2.inRange(hsv, (0, 50, 50), (10, 255, 255)) | cv2.inRange(hsv, (160, 50, 50), (180, 255, 255))
```

## 202331034_Nabila Ridha Adman

```python
masks = [
    ('NONE', np.zeros_like(mask_blue)),
    ('BLUE', mask_blue),
    ('RED-BLUE', (mask_red | mask_blue).astype(np.uint8)*255),
    ('RED-GREEN-BLUE', (mask_red | mask_green | mask_blue).astype(np.uint8)*255)]
```

## 202331034_Nabila Ridha Adman

```python
plt.figure(figsize=(12, 8))

for i, (title, mask) in enumerate(masks, 1):
    plt.subplot(2, 2, i)
    plt.imshow(mask, cmap='gray')
    plt.title(title)
    plt.axis('off')
    
plt.tight_layout()
plt.show()
```
